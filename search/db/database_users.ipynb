```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
conn = sqlite3.connect("data.db")

df = pd.read_sql("select * from  userstable", con=conn)

conn.close()
```

```python
df['Statut'] = 'utilisateur'
df['Statut'][8] = 'administrateur'
dummy_statut = pd.get_dummies(df['Statut']) # le code encode le statut du client.
df =  pd.merge(left=df,right=dummy_statut,left_index=True,right_index=True,)
df = df.drop(["Statut"], axis = 1)
df
```

```python
df.to_csv('Users.csv',index = False) #pour sauvegarder la nouvelle database sous forme csv
```
